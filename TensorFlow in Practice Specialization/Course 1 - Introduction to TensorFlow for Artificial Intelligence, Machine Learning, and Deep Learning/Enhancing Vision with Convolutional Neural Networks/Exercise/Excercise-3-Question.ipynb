```python
# GRADED FUNCTION: train_mnist_conv
def train_mnist_conv():
    # Please write your code only where you are indicated.
    # please do not remove model fitting inline comments.
    
    # init values
    epochs = 20
    min_acc = 0.998
    acc_string = "\nReached 99.8% accuracy so cancelling training!"

    # YOUR CODE STARTS HERE
    # implement callback
    class Callback(tf.keras.callbacks.Callback):
        def on_epoch_end(self, epoch, log={}):
            if(log.get("acc")> min_acc):
                print(acc_string)
                self.model.stop_training = True      
    # YOUR CODE ENDS HERE

    mnist = tf.keras.datasets.mnist
    (training_images, training_labels), (test_images, test_labels) = mnist.load_data(path=path)
    
    # YOUR CODE STARTS HERE
    # preprocess images
    training_images = training_images.reshape(60000, 28, 28, 1)
    test_images = test_images.reshape(10000, 28, 28, 1)
    # scale images
    training_images = training_images / 255
    test_images = test_images / 255
    # YOUR CODE ENDS HERE
    
    
    callback = Callback()
    
    model = tf.keras.models.Sequential([
        # YOUR CODE STARTS HERE
        tf.keras.layers.Conv2D(filters = 32, kernel_size = (3,3), 
                               activation = 'relu', input_shape = (28,28,1)), 
        tf.keras.layers.MaxPool2D(pool_size = (2,2)),
        tf.keras.layers.Flatten(), # input layer - don´t need to specify
        tf.keras.layers.Dense(units = 128, activation = "relu"), # hidden layer
        tf.keras.layers.Dense(units = 10, activation = "softmax") # output layer - 10 categories
            # YOUR CODE ENDS HERE
    ])

    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    model.summary()
    # model fitting
    history = model.fit(
        # YOUR CODE STARTS HERE
        x = training_images,
        y = training_labels,
        epochs = epochs,
        callbacks = [callback]
        # YOUR CODE ENDS HERE
    )
    # model fitting
    return history.epoch, history.history['acc'][-1]

```

```python
nr_epochs, train_acc = train_mnist_conv()
```

```python
print(nr_epochs)
print(train_acc)
```
